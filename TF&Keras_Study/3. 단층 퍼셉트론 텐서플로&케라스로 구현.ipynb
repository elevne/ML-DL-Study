```python
import numpy as np
import tensorflow as tf
```

```python
# 웨이트와 바이어스 초기화
w = tf.Variable(tf.zeros([2, 1])) # 2차원
b = tf.Variable(tf.zeros([1]))
```

```python
# 수식 정의하기
def y(x):
    return sigmoid(np.dot(w, x) + b)
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
import tensorflow.compat.v1 as tf
tf.disable_v2_behavior()

x = tf.placeholder(tf.float32, shape=[None, 2])
t = tf.placeholder(tf.float32, shape=[None, 1])

tf.disable_v2_behavior()
y = tf.nn.sigmoid(tf.matmul(x, w) + b) # matmul은 행렬곱
```

```python
# 오차함수 구현해두기
cross_entropy = -tf.reduce_sum(t*tf.log(y) + (1-t)*tf.log(1-y))
```

```python
# 경사하강법 구현하기
train_step = tf.train.GradientDescentOptimizer(0.1).minimize(cross_entropy)
```

```python
# 뉴런의 발화여부 정하는 기준~~
correct_prediction = tf.equal(tf.to_float(tf.greater(y, 0.5)), t)
```

```python
# 학습용 데이터 만들기
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [1]])
```

```python
# 텐서플로에서는 데이터를 취급하는 방법인 세션이라는 것 안에서 계산 실행~~
tf.disable_v2_behavior()
init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init) # 모델을 정의할 때 선언했던 변수와 식을 초기화하는 것~
```

```python
# 학습 구현해보기
for epoch in range(200):
    sess.run(train_step, feed_dict={
        x: X, t: Y
    })
```

```python
# 적절히 분류한 것들 보기
classified = correct_prediction.eval(session=sess, feed_dict={x : X, t : Y})
classified
```

```python
# 각 입력에 대한 출력 확률 보기
prob = y.eval(session=sess, feed_dict={x:X, t:Y})
prob
```

```python
# tf.Variable()로 정의한 변수의 값은 .eval이 아니라 sess.run()으로 구할 수 있음
print('w :',sess.run(w))
print('b :', sess.run(b))
```

#### 이번에는 Keras로 구현해보기

```python
import numpy as np
from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.optimizers import SGD

model = Sequential([Dense(input_dim=2, units=1), Activation('sigmoid')])
```

```python
# 경사하강법 구현
model.compile(loss='binary_crossentropy', optimizer=SGD(lr=0.1))
```

```python
# 학습용 데이터 만들기
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0], [1], [1], [1]])
```

```python
# 학습하기
model.fit(X, Y, epochs=200, batch_size=1)
```

```python
# 학습 결과 확인해보기
classes = model.predict(X, batch_size=1)
classes
```
